### 5. concatenate()

```python
array1 = np.concatenate([array1, array2, array3, ..., arrayN],axis = 0)
axis = 0  >> Rows
axis = 1  >> columns
```

```python
class0 = np.zeros(10)
class1 = np.ones(10)
class2 = np.full(10,2)
print(class2)
new_array = np.concatenate([class0, class1, class2])
new_array
```
